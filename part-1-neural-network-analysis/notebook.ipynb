```python
# ==========================================
# IMPORTING REQUIRED LIBRARIES
# ==========================================

import pandas as pd
import numpy as np

# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Sklearn libraries 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# TensorFlow / Keras libraries for building neural networks
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Ignore unnecessary warnings
import warnings
warnings.filterwarnings('ignore')


# ==========================================
# TASK 1: Dataset Understanding 
# ==========================================

# Loading the dataset
df = pd.read_csv('customer_churn_nn.csv')

# Displaying the first few rows of the dataset
print("\n========== First 5 rows of the dataset ==========")
print(df.head())

# Checking number of ROWS and COLUMNS
print("\n========== Dataset Shape ==========")
print("Rows and Columns:", df.shape)

# Checking column names and types of input features
print("\n========== Column Names and Data Types ==========")
print("Column Names: \n" + str(df.columns))
print("\nData Types: \n" + str(df.dtypes))

# Target Variable Distribution
# The target variable is 'churn', which indicates whether a customer has churned = (1) or not(retained) = (0).
print("\n========== Target Variable Distribution ==========")
print(df['churn'].value_counts())

# Checking for missing values
print("\n========== Missing Values ==========")
print(df.isnull().sum())

# Basic Statistical Summary
print("\n========== Statistical Summary ==========")
print(df.describe())

# Distribution of the target Variable
plt.figure(figsize=(6, 4))

sns.countplot(x='churn', data=df)

plt.title("Distribution of Target Variable (Churn)")
plt.xlabel("Churn")
plt.ylabel("Count")

plt.savefig("results/target_distribution.png")

plt.show()

# ==========================================
# TASK 2: Data Preprocessing
# ==========================================

# REMOVING IDENTIFIER COLUMNS
# customer_id is just an identifier
# It does not provide any useful information for the model and can be removed.

df = df.drop('customer_id', axis=1)
print("\n===== DATASET AFTER REMOVING customer_id =====")
print(df.head())


# HANDLING MISSING VALUES
# Separate numerical and categorical columns

numerical_cols = df.select_dtypes(include=np.number).columns
categorical_cols = df.select_dtypes(include='object').columns

# Filling missing values in numerical columns with mean
for col in numerical_cols:
    df[col].fillna(df[col].mean(), inplace=True)

# Filling missing values in categorical columns with mode
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Checking for missing values after handling
print("\n========== MISSING VALUES AFTER HANDLING ==========")
print(df.isnull().sum())


# ENCODING CATEGORICAL VARIABLES
# Neural networks only understand numbers, So Categorical columns must be converted into numerical values.

label_encoder = LabelEncoder()

for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])

print("\n========== DATASET AFTER ENCODING CATEGORICAL VARIABLES ==========")
print(df.head())


# SPLITTING THE DATA INTO TRAINING AND TESTING SETS
# X -> Input features
# y -> Target variable

X = df.drop('churn', axis=1)
y = df['churn']


# FEATURE SCALING
# Scaling impves neural network performance, StandardScaler converts values into similar scale

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# Splitting the data into training and testing sets (80% train, 20% test)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, 
    y, 
    test_size=0.2, 
    random_state=42,
    stratify=y
)

print("\n========== SHAPE OF TRAINING AND TESTING SETS ==========")

print("Training Data Shape:", X_train.shape)
print("Testing Data Shape:", X_test.shape)


# ==========================================
# TASK 3: Building the Neural Network Model
# ==========================================

# Creating a Sequential model
model = Sequential()

# Adding input layer and first hidden layer
model.add(
    Dense(
        units=16,                       # Number of neurons
        activation='relu',              # Activation function
        input_shape=(X_train.shape[1],) # Number of input features
    )
)

# Adding second hidden layer
model.add(
    Dense(  
        units=8, 
        activation='relu'
    )
)

# Adding output layer
model.add(
    Dense(
        units=1, 
        activation='sigmoid'  # Sigmoid activation for binary classification
    )
)

# Compiling the model
model.compile(
    optimizer=Adam(learning_rate=0.001),  # Optimizer
    loss='binary_crossentropy',             # Loss function for binary classification
    metrics=['accuracy']                    # Evaluation metric
)

# Displaying the model architecture
print("\n========== NEURAL NETWORK MODEL SUMMARY ==========")
model.summary()


# ==========================================
# TASK 4: Training And Evaluation
# ==========================================

# Since churned customers are very few, we assign higher importance to class 1
from sklearn.utils.class_weight import compute_class_weight

class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)

class_weights = dict(enumerate(class_weights))

print("Class Weights:", class_weights)

# Training the model
history = model.fit(
    X_train, 
    y_train, 
    epochs=50, 
    batch_size=32, 
    validation_split=0.2, 
    class_weight=class_weights,
    verbose=1
)

# Plot Training Accuracy and Loss

plt.figure(figsize=(12, 5))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title("Training VS Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Training VS Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.tight_layout()

# Save training graphs
plt.savefig("results/evaluation_outputs.png")

plt.show()

# Evaluating the model on the test dataset

test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print("\n========== TEST RESULTS ==========")
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")

# Making Predictions
y_pred_prob = model.predict(X_test)       # Predicting probabilities for the test set
y_pred = (y_pred_prob > 0.5).astype(int).flatten()  # Converting probabilities to binary predictions


# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 4))

sns.heatmap(
    cm, 
    annot=True, 
    fmt='d', 
    cmap='Blues'
)

plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.savefig("results/confusion_matrix.png")

plt.show()

# Classification Report
print("\n========== CLASSIFICATION REPORT ==========")
print(classification_report(y_test, y_pred))


# INTERPRETATION 

print("\n=============== MODEL INTERPRETATION ===============")

print("1. The model achieved a test accuracy of {:.2f}%.".format(test_accuracy * 100))
print("2. Since the dataset is highly imbalanced, recall and F1-score are also important evaluation metrics.")
print("3. The classification report provides precision, recall, and F1-score for both classes (churned and retained).")
print("4. The model seems to perform well, but further tuning and experimentation may be needed to improve performance.")

print("""
The neural network learned meaningful patterns from the dataset.

Training accuracy increased steadily while training loss decreased,
which indicates successful learning.

The confusion matrix shows how many predictions were correct and incorrect.

A higher accuracy means the model can correctly identify customer churn effectively.
""")


# ==========================================
# TASK 5: Hyperparameter Experimentation
# ==========================================

# Creating Function For Experiments 
def build_and_train_model(neurons, learning_rate, batch_size, epochs):
    
    model = Sequential()          # Creating model
    
    # Hidden layer 1
    model.add(
        Dense(
            neurons,
            activation='relu',
            input_dim=X_train.shape[1]
        )
    )

    # Hidden layer 2
    model.add(
        Dense(
            neurons // 2, 
            activation='relu'
        )
    )

    # Output layer
    model.add(
        Dense(
            1, 
            activation='sigmoid'
        )
    )

    # Compiling the model
    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )   

    # Training the model
    model.fit(
        X_train, 
        y_train, 
        epochs=epochs, 
        batch_size=batch_size,
        class_weight=class_weights,
        verbose=0
    )

    # Evaluating the model
    loss, accuracy = model.evaluate(X_test, y_test, verbose=0)

    return accuracy 

# RUNNING MULTIPLE EXPERIMENTS
experiment_results = []

# Experiment 1: Baseline Model Configuration
acc1 = build_and_train_model(
    neurons=16, 
    learning_rate=0.001, 
    batch_size=32, 
    epochs=30
)
    
experiment_results.append([
    "Experiment 1",
    16,
    0.001,
    32,
    30,
    acc1
])

# Experiment 2: Increasing Neurons, Reduced Batch Size, Increased Epochs
acc2 = build_and_train_model(
    neurons=32, 
    learning_rate=0.001, 
    batch_size=16, 
    epochs=40
)

experiment_results.append([
    "Experiment 2",
    32,
    0.001,
    16,
    40,
    acc2
])

# Experiment 3: Increased Neurons, Higher Learning Rate, More Epochs
acc3 = build_and_train_model(
    neurons=64,
    learning_rate=0.01,
    batch_size=32,
    epochs=50
)

experiment_results.append([
    "Experiment 3",
    64,
    0.01,
    32,
    50,
    acc3
])

# Creating Comparison Table
comparison_df = pd.DataFrame(
    experiment_results,
    columns=[
        "Experiment",
        "Neurons in Hidden Layer",
        "Learning Rate",
        "Batch Size",
        "Epochs",
        "Test Accuracy"
    ]
)

print("\n=============== HYPERPARAMETER EXPERIMENTATION RESULTS ===============")

print(comparison_df)

# Saving Comparison Table to CSV
# Save comparison table inside results folder

comparison_df.to_csv(
    "results/model_comparison_table.csv",
    index=False
)

print("Comparison table saved successfully.")

# ==========================================
# Task 6: FINAL REFLECTION
# ==========================================

print("\n=============== FINAL REFLECTION ===============")

# ---------------
# Question 1
# ---------------
print("1. WHAT ROLE DO WEIGHTS AND BIASES PLAY?\n")

print("""
Weights determine the importance of input features in prediction.

Biases help shift the activation function and improve flexibility of learning.

During training, the neural network continuously updates weights and biases
to reduce prediction error.
""")

# ---------------
# Question 2
# ---------------
print("2. WHY IS ACTIVATION FUNCTION REQUIRED?\n")

print("""
Activation functions introduce non-linearity into the neural network.

Without activation functions, the network would behave like a simple
linear model and would not be able to learn complex patterns.
""")

# ---------------
# Question 3
# ---------------
print("3. WHAT HAPPENS WHEN LEARNING RATE IS TOO HIGH OR TOO LOW?\n")

print("""
If learning rate is too high:
- The model may overshoot the optimal solution
- Training becomes unstable

If learning rate is too low:
- Training becomes very slow
- Model may take too much time to converge
""")

# ---------------
# QUESTION 4
# ---------------
print("4. DID THE MODEL SHOW UNDERFITTING OR OVERFITTING?\n")

print("""
The model showed slight signs of overfitting because training loss continued decreasing while validation loss became more stable after several epochs.

The dataset was also highly imbalanced, which made prediction of churn customers more difficult.

You can observe this behavior from the training graphs.
""")

# =====================================================================================================
# END OF ASSIGNMENT
# =====================================================================================================
```
